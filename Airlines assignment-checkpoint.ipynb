### Hierarchy Clustring

```python
from sklearn.preprocessing import normalize
Airline_norm=pd.DataFrame(normalize(airline_data),columns=airline_data.columns)
Airline_norm
```

```python
dendrogram=sch.dendrogram(sch.linkage(Airline_norm,method='average'))
```

```python
hc=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
```

```python
y=pd.DataFrame(hc.fit_predict(Airline_norm),columns=['clusters'])
```

```python
airline_data['clustering']=hc.labels_
```

```python
airline_data.groupby('clustering').agg(['mean']).reset_index()
```

```python
plt.figure(figsize=(10, 6))  
plt.scatter(airline_data['clustering'],airline_data['Balance'], c=hc.labels_)
plt.show()
```

### K-Mean Clustering

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
Airline_norm2=StandardScaler().fit_transform(airline_data)
Airline_norm2
```

```python
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(Airline_norm2)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

### Build Cluster algorithm using K=6

```python
clusters6=KMeans(6,random_state=30).fit(Airline_norm2)
clusters6
```

```python
clusters6.labels_
```

```python
Airline_norm3 = airline_data.copy()
Airline_norm3 ['clustersing'] = clusters6.labels_
Airline_norm3.head()
```

```python
Airline_norm3.groupby('clustersing').agg(['mean']).reset_index()
```

```python
plt.figure(figsize=(10, 6))  
plt.scatter(Airline_norm3['clustersing'],Airline_norm3['Qual_miles'], c=clusters6.labels_) 
plt.show()
```

### DBSCAN Clustering

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
Airline_norm4=StandardScaler().fit_transform(airline_data)
Airline_norm4
```

```python
dbscan=DBSCAN(eps=1,min_samples=6)
dbscan.fit(Airline_norm4)
```

```python
dbscan.labels_
```

```python
airline_data['clusters']=dbscan.labels_
airline_data
```

```python
airline_data.groupby('clusters').agg(['mean']).reset_index()
```

```python
plt.figure(figsize=(10, 6))  
plt.scatter(airline_data['clusters'],airline_data['Balance'], c=dbscan.labels_) 
plt.show()
```
