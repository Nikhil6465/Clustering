```python
df1 = crime_data.drop(labels='States',axis=1)
```

### Hierarchy Clustring

```python
from sklearn.preprocessing import normalize
crime_data_norm=pd.DataFrame(normalize(df1),columns=df1.columns)
crime_data_norm
```

```python
# Create dendrograms
dendrogram=sch.dendrogram(sch.linkage(crime_data_norm,method='average'),p=50)
```

```python
hc=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
```

```python
y=pd.DataFrame(hc.fit_predict(crime_data_norm),columns=['clusters'])
y
```

```python
crime_data['Clustering']=hc.labels_
```

```python
crime_data
```

```python
crime_data.groupby('Clustering').agg(['mean']).reset_index()
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(crime_data['Clustering'],crime_data['Murder'], c=hc.labels_) 
plt.show()
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(crime_data['Clustering'],crime_data['Assault'], c=hc.labels_) 
plt.show()
```

### K-Mean Clustering

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
crime_norm2=StandardScaler().fit_transform(df1)
crime_norm2
```

```python
# within-cluster sum-of-squares criterion 
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(crime_norm2)
    wcss.append(kmeans.inertia_)
```

```python
# Plot K values range vs WCSS to get Elbow graph for choosing K (no. of clusters)
plt.plot(range(1,11),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

### Build Cluster algorithm using K=4

```python
# Cluster algorithm using K=4
clusters4=KMeans(4,random_state=30).fit(crime_norm2)
clusters4
```

```python
clusters4.labels_
```

```python
# Assign clusters to the data set
crime_norm3 = crime_data.copy()
crime_norm3 ['Clustering'] = clusters4.labels_
crime_norm3
```

```python
# Compute the centroids for K=4 clusters with 4 variables
clusters4.cluster_centers_
```

```python
# Group data by Clusters (K=4)
crime_norm3.groupby('Clustering').agg(['mean']).reset_index()
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(crime_norm3['Clustering'],crime_norm3['Murder'], c=clusters4.labels_) 
plt.show()
```

### DBSCAN Clustering

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crime_norm4=StandardScaler().fit_transform(df1)
crime_norm4
```

```python
# DBSCAN Clustering
dbscan=DBSCAN(eps=1,min_samples=4)
dbscan.fit(crime_norm4)
```

```python
dbscan.labels_
```

```python
crime_data['Clustering']=dbscan.labels_
```

```python
crime_data.groupby('Clustering').agg(['mean']).reset_index()
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(crime_data['Clustering'],crime_data['Assault'], c=dbscan.labels_)
plt.show()
```
